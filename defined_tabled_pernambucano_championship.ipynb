```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns

print("\n------------------CAMPEONATO PERNAMBUCANO 2025------------------")
print('\n Regras do campeonato:')
print('4 primeiros vão para o mata-mata:')
print('2 ultimos são rebaixados:')
print('\n Tabela do campeonato:')
print('\n ')


uploaded = files.upload()
data = pd.read_csv ('pernambucano.csv', encoding='iso-8859-1')

print(data)

df1=pd.DataFrame(data)

melhor = df1.iloc[0]['Clube']
pior = df1.iloc[0]['Clube']
semis = df1.head(4)['Clube'].tolist()
rebaixados = df1.tail(3)['Clube'].tolist()
linha_gols = df1.loc[df1['Gols Marcados'].idxmax()]
time_artilheiro = linha_gols['Clube']
time_gols = linha_gols['Gols Marcados']
defesa_ruim = df1.loc[df1['Gols Marcados'].idxmin()]
time_ruim = defesa_ruim['Clube']
gols_ruim = defesa_ruim['Gols Marcados']

print('\nO melhor time do campeonato foi o', melhor)
print('\nO pior time do campeonato foi o', pior)
print("\nClassificados para o mata-mata:", ', '.join(semis))
print('\nOs times rebaixados para a serie A2 do campeonato foram:', ', '.join(rebaixados))
print('\nO melhor ataque foi do time', time_artilheiro, 'com', time_gols, 'gols')
print('\nA defesa mais vazada foi do time', time_ruim, 'com', gols_ruim, 'tomados')


sns.set_theme(style="whitegrid")
#graficodebarra
plt.figure(figsize=(12, 4))
sns.barplot(x='Clube', y='Pontos', data=df1, palette='viridis')
plt.title('Pontuação Campeonato Pernambucano')
plt.ylabel('Pontos')
plt.xlabel('Clube')
plt.tight_layout()
plt.show()

#graficodepontos
plt.figure(figsize=(12, 6))
sns.lineplot(x='Clube', y='Gols Marcados', data=df1, label='Gols Feitos', linestyle='-', color='green', linewidth=5)
sns.lineplot(x='Clube', y='Gols Contra', data=df1, label='Gols Tomados' ,linestyle='-', color='red', linewidth=5)
plt.title('Gols pro vs Gols contra')
plt.show()


```
